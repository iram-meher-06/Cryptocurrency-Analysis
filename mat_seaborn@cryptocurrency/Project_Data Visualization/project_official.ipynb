```python
from jupyterthemes import jtplot
jtplot.style(theme = 'monokai', context = 'notebook', ticks = True, grid = False) 

'''In this project, we will analyze cryptocurrency prices such as Bitcoin (BTC), Ethereum (ETH), Litecoin (LTC), Cardano (ADA) and Ripple (XRP). 
Cryptocurrency is a decentralized digital currency that uses cryptography to secure transactions and do not have a centralized issuing authority
(Government or banks). Using the “crypto_daily_prices.csv” and “crypto_daily_returns.csv” datasets , we perform the following: 
1. Import both datasets using Pandas.
2. Using Matplotlib, plot lineplots that display crypto daily prices on one single figure.
3. Using Matplotlib, plot crypto daily prices on multiple subplots.
4. Using Matplotlib, plot the scatterplot between Bitcoin and Ethereum daily returns. 
5. Using Seaborn, plot similar scatterplot between bitcoin and Ethereum daily returns. 
6. Assume that you became bullish on XRP and decided to allocate 60% of your assets in it. You also decided to equally divide the rest of your assets in 
other coins (BTC, LTC, ADA, and ETH). Using Matplotlib, plot a pie chart that shows these allocations. Use 'explode’ attribute to increase the separation 
between BTC and the rest of the portfolio.
7. Using Matplotlib, plot the histogram for ETH returns using 30 bins with red color.
8. Using Seaborn, plot a heatmap that shows the correlations between crypto daily returns.'''

# 1. Import both datasets using Pandas.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

investments_df= pd.read_csv('crypto_daily_prices.csv')
investments_df
```

```python
# 2. Using Matplotlib, plot lineplots that display crypto daily prices on one single figure.

investments_df.plot(x='Date' , y=[ 'BTC-USD Price','ETH-USD Price','LTC-USD Price' ] , label=[ 'Bitcoin Price',' Ethereum Price', 'LiteCoin Price' ], figsize=(10,5), linewidth= 2)
plt.ylabel('Price[$]')
plt.grid()
plt.title('Crypto Daily Prices')
plt.legend(loc='upper right')
```

```python
investments_df.plot(x='Date', title='Crypto Prices', subplots=True, grid= True, figsize=(15,25))
```

```python
daily_return_df= pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
# # 4. Using Matplotlib, plot the scatterplot between Bitcoin and Ethereum daily returns.

daily_return_df.plot.scatter('BTC','ETH',color='r', figsize=(8,5))
```

```python
# 5. Using Seaborn, plot similar scatterplot between bitcoin and Ethereum daily returns. 

import seaborn as sns

plt.figure(figsize=(8,5))
#sns.scatterplot(daily_return_df['BTC'],daily_return_df['ETH'])
sns.scatterplot(data=daily_return_df,x='BTC', y='ETH')
```

```python
# 6. Assume that you became bullish on XRP and decided to allocate 60% of your assets in it. You also decided to equally divide the rest of your assets in 
#other coins (BTC, LTC, ADA, and ETH). Using Matplotlib, plot a pie chart that shows these allocations. Use 'explode’ attribute to increase the separation 
#between XRP and the rest of the portfolio.

# Define a dictionary with all crypto allocation in a portfolio
# Note that total summation = 100% 

my_dict= { 'allocation %': [10,10,10,60,10]}
explode_l= (0,0,0,0.2,0)

cryto_df=pd.DataFrame(data= my_dict, index=['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
cryto_df.plot.pie(y= 'allocation %', explode= explode_l, figsize=(8,8))

plt.title('Crypto portfolio pie chart')
```

```python
# 7. Using Matplotlib, plot the histogram for ETH returns using 30 bins with red color.
# alpha = 0.5 means that the red bars in your histogram will be 50% opaque and 50% transparent
#  An alpha of 1 would be completely solid glass, while an alpha of 0 would be completely invisible
mu= daily_return_df['ETH'].mean()
sigma= daily_return_df['ETH'].std()

nbins=40

plt.figure(figsize=(12,6))
daily_return_df['ETH'].plot.hist(bins=30,alpha= 0.5,  facecolor='c')
plt.grid()
plt.title('Histogram: mu= '+ str(mu)+"   sigma: "+str(sigma))
plt.ylabel('Probability')
```
