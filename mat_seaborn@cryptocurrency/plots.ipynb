```python
daily_return_df= pd.read_csv('stock_daily_return.csv')
daily_return_df
```

```python
x= daily_return_df['FB']
y= daily_return_df['TWTR']
plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.grid()
```

```python
value_l = [20,40,30,10]
color_l= ['c','r','y','m']
label_l= ['APL','GGLE','TSLA','AMZN']
explode_l=[0,0,0.1,0.1]

plt.figure(figsize=(5,5))
plt.pie(value_l, colors=color_l, labels= label_l, explode= explode_l)
plt.title('STOCK PORTFOLIO')
#plt.legend()
```

```python
# sd-dispersion away from the mean
# A histogram represents data using bars of various heights. 
# Each bar groups numbers into specific ranges. 
# Taller bars show that more data falls within that specific range.
mu= daily_return_df['FB'].mean()
sigma= daily_return_df['FB'].std()

nbins= 40

plt.figure(figsize=(7,7))
plt.hist(daily_return_df['FB'],nbins, facecolor='r')
plt.grid()
plt.title("Histogram: mu = "+ str(mu)+" Sigma = "+str(sigma))
```

```python
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html

# import libraries 
import seaborn as sns # Statistical data visualization
```

```python
from sklearn.datasets import load_breast_cancer   #target is output
cancer= load_breast_cancer()
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
#The first argument provides the data that will populate the DataFrame, and the columns argument specifies the names for each column.
#  np.c_[]:  Special index object that performs column-wise concatenation. 
#This is a convenient way to combine arrays (or array-like structures) as columns into a single array.
#np.c_[] is stacking these as separate columns.
#it assigns column names by taking the original feature names and adding "target" as the name for the last column (which contains the target variable).

cancer_df= pd.DataFrame(np.c_[ cancer['data'],cancer['target'] ] , columns= np.append(cancer['feature_names'], ['target']))
```

```python
cancer_df.head()   #tail- bottom 5, total- 568 samples
```

```python
#Scatter plot, sns:

plt.figure(figsize=(10,10))
sns.scatterplot(x='mean area',y='mean smoothness',hue='target',data= cancer_df)
#to specify which dpts belong to malignant class & benign class (have cancer or not), specify color(hue)
#The purpose of the hue argument in sns.scatterplot() is to visually distinguish data points based on a categorical variable.
```

```python
# access to a new dataset- use pairplot to get insights-for a (**Classification type problem)
#check whther these dpts are separable from e.o or not(if grp green ones and blue ones sep- then spearable)
# 1. pairplot:

sns.pairplot(cancer_df, hue='target',vars=['mean area','mean texture','mean radius','mean perimeter','mean smoothness'])
```

```python
#heatmap:
#correlation bw two variables or among features here: white- highly correlated, -ve value(darker acc to scale)-> less correlated

plt.figure(figsize=(30,30))
sns.heatmap(cancer_df.corr(), annot=True)
```

```python
# 3. displot: updates to displot

# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 



sns.displot(cancer_df['mean radius'], bins= 25, color='b')
```
